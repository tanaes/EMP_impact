```python
import pandas as pd
import scholar as gs
from StringIO import StringIO
```

```python
emp_studies = pd.read_csv('./emp_studies.csv', index_col=0)
emp_studies
```

```python
def csv_str(querier, header=False, sep='|'):
    output = ''
    articles = querier.articles
    for art in articles:
        result = art.as_csv(header=header, sep=sep)
        output += gs.encode(result)
        header = False
    return(output)
```

```python
def doi_search(doi):

    querier = gs.ScholarQuerier()
    settings = gs.ScholarSettings()

    settings.set_citation_format(gs.ScholarSettings.CITFORM_BIBTEX)
    querier.apply_settings(settings)

    query = gs.SearchScholarQuery()
    query.set_phrase(doi)
    querier.send_query(query)
    
    output = pd.read_csv(StringIO(csv_str(querier, header=True)), sep='|', header=0)
    
    return(output['title'][0], output['year'][0], output['num_citations'][0])
```

```python
emp_study_idx = []
emp_study_nums = []
emp_dois = []
emp_titles = []
emp_years = []
emp_cites = []
```

```python
# had to perform this in chunks to not trip Google bot detectors
these_studies = emp_studies[23:52]

for study in these_studies.iterrows():
    dois = study[1]['doi']
    
    if pd.isnull(dois):
        continue    
    if dois.startswith('Missing'):
        continue
    
    doi_list = dois.split(',')
    
    for doi in doi_list:
        doi = doi.strip()
        
        title, year, cites = doi_search(doi)

        print('Title: {0}\nYear: {1}\nCites: {2}\n'.format(title, year, cites))

        emp_study_idx.append(study[0])
        emp_study_nums.append(study[1]['study_id'])
        emp_dois.append(doi)
        emp_titles.append(title)
        emp_years.append(year)
        emp_cites.append(cites)
    
```

```python
emp_citation_table = pd.DataFrame({'study_idx': emp_study_idx, 'emp_study_num': emp_study_nums, 'doi': emp_dois, 'title': emp_titles, 'year': emp_years, 'cites': emp_cites})
```
